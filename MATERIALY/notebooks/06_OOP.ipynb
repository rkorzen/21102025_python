## 1. Wprowadzenie do programowania obiektowego

- OOP to paradygmat, w ktÃ³rym modelujemy problem jako wspÃ³Å‚pracujÄ…ce ze sobÄ… obiekty.
- Kluczowe pojÄ™cia:
  - **obiekt**: konkretna instancja klasy
  - **klasa**: przepis/typ definiujÄ…cy strukturÄ™ (atrybuty) i zachowanie (metody)
  - **atrybut**: pole danych przechowywane w obiekcie lub klasie
  - **metoda**: funkcja zdefiniowana wewnÄ…trz klasy
  - **instancja**: konkretny obiekt utworzony na podstawie klasy
- **Enkapsulacja**: ukrywanie szczegÃ³Å‚Ã³w implementacji za czytelnym interfejsem (metodami/properties).

ðŸ§© Zadanie: UtwÃ³rz klasÄ™ `Car` z atrybutami `brand`, `model`, `year` i metodÄ… `start()` wypisujÄ…cÄ… komunikat.


```
# PrzykÅ‚adowa realizacja zadania (moÅ¼esz zmodyfikowaÄ‡):
class Car:
    def __init__(self, brand: str, model: str, year: int):
        self.brand = brand
        self.model = model
        self.year = year

    def start(self) -> None:
        # Prosta metoda pokazujÄ…ca zachowanie obiektu
        print(f"{self.brand} {self.model} ({self.year}) -> silnik uruchomiony!")

# KrÃ³tki test
car = Car("Toyota", "Corolla", 2020)
car.start()
```

## 2. Definicja klasy i instancji

- SÅ‚owo kluczowe `class` definiuje klasÄ™.
- Konstruktor `__init__` inicjalizuje nowo tworzonÄ… instancjÄ™.
- `self` to referencja do bieÅ¼Ä…cej instancji (jak `this` w innych jÄ™zykach).
- Atrybuty klasy sÄ… wspÃ³Å‚dzielone przez wszystkie instancje, atrybuty instancji â€” specyficzne dla obiektu.

ðŸ“˜ PrzykÅ‚ad: prosta klasa `Person` z metodami `greet()` i `is_adult()`.


```
class Person:
    species = "Homo sapiens"  # atrybut klasowy (wspÃ³lny)

    def __init__(self, name: str, age: int):
        self.name = name        # atrybut instancyjny
        self.age = age          # atrybut instancyjny

    def greet(self) -> None:
        print(f"CzeÅ›Ä‡, jestem {self.name} (gatunek: {Person.species})")

    def is_adult(self) -> bool:
        return self.age >= 18

p = Person("Jan", 22)
p.greet()
print("DorosÅ‚y?", p.is_adult())
```

## 3. Enkapsulacja (Hermetyzacja)

- Konwencje nazewnicze:
  - `_name` â€” atrybut â€žchronionyâ€ (umowny, sygnaÅ‚ dla programistÃ³w)
  - `__name` â€” name mangling (utrudnia przypadkowy dostÄ™p, nie jest to peÅ‚na prywatnoÅ›Ä‡)
- Getter/Setter moÅ¼na pisaÄ‡ jawnie lub uÅ¼yÄ‡ `@property`, aby oferowaÄ‡ interfejs atrybutu z walidacjÄ….

ðŸ“˜ PrzykÅ‚ad: `BankAccount` z prywatnym saldem i kontrolowanym dostÄ™pem przez `@property`.

ðŸ§© Zadanie: Dodaj metody `deposit(amount)` i `withdraw(amount)`, ktÃ³re modyfikujÄ… saldo tylko przy poprawnych danych.


```
class BankAccount:
    def __init__(self, owner: str, initial_balance: float = 0.0):
        self.owner = owner
        self.__balance = float(initial_balance)  # "prywatne" przez name mangling

    @property
    def balance(self) -> float:
        # Tylko odczyt przez property (na razie bez settera)
        return self.__balance

    def deposit(self, amount: float) -> None:
        # Walidacja kwoty i modyfikacja stanu obiektu
        if amount <= 0:
            raise ValueError("Kwota depozytu musi byÄ‡ dodatnia")
        self.__balance += amount

    def withdraw(self, amount: float) -> None:
        if amount <= 0:
            raise ValueError("Kwota wypÅ‚aty musi byÄ‡ dodatnia")
        if amount > self.__balance:
            raise ValueError("Brak wystarczajÄ…cych Å›rodkÃ³w")
        self.__balance -= amount

acct = BankAccount("Ala", 100)
acct.deposit(50)
try:
    acct.withdraw(200)
except ValueError as e:
    print("BÅ‚Ä…d:", e)
print("Saldo:", acct.balance)
```

## 4. Dziedziczenie

- Pozwala ponownie uÅ¼ywaÄ‡ i rozszerzaÄ‡ istniejÄ…ce klasy.
- `super()` umoÅ¼liwia wywoÅ‚anie logiki klasy bazowej.
- Python wspiera teÅ¼ dziedziczenie wielokrotne (ostroÅ¼nie, patrz MRO poniÅ¼ej).

ðŸ“˜ PrzykÅ‚ad: `Animal` â†’ `Dog` z nadpisanÄ… metodÄ… `speak()`.

ðŸ§© Zadanie: UtwÃ³rz hierarchiÄ™ `Vehicle` â†’ `Car` â†’ `ElectricCar`. Niech `ElectricCar` nadpisuje `fuel_type()` zwracajÄ…cÄ… "electric".


```
class Animal:
    def speak(self) -> str:
        return "(cisza)"

class Dog(Animal):
    def speak(self) -> str:  # nadpisanie metody
        return "Woof!"

a = Animal()
d = Dog()
print(a.speak())
print(d.speak())

# PrzykÅ‚adowe rozwiÄ…zanie zadania
class Vehicle:
    def fuel_type(self) -> str:
        return "unknown"

class Car(Vehicle):
    def fuel_type(self) -> str:
        return "gasoline"

class ElectricCar(Car):
    def fuel_type(self) -> str:
        return "electric"

print(ElectricCar().fuel_type())
```

## 5. Polimorfizm

- Ten sam interfejs (nazwy metod), rÃ³Å¼ne implementacje.
- W Pythonie czÄ™sto uÅ¼ywamy duck typing: â€žjeÅ›li coÅ› ma metodÄ™ `area()`, traktujemy to jak figurÄ™ geometrycznÄ…â€.

ðŸ“˜ PrzykÅ‚ad: `Shape`, `Circle`, `Square` â€” kaÅ¼da ma metodÄ™ `area()` dziaÅ‚ajÄ…cÄ… inaczej.

ðŸ§© Zadanie: Napisz funkcjÄ™ `print_area(shape)`, ktÃ³ra wywoÅ‚uje `shape.area()` bez sprawdzania typu.


```
import math

class Shape:
    def area(self) -> float:
        raise NotImplementedError

class Circle(Shape):
    def __init__(self, r: float):
        self.r = r
    def area(self) -> float:
        return math.pi * self.r ** 2

class Square(Shape):
    def __init__(self, a: float):
        self.a = a
    def area(self) -> float:
        return self.a * self.a

def print_area(shape) -> None:
    # Duck typing: zakÅ‚adamy, Å¼e obiekt ma metodÄ™ area()
    print("Area:", shape.area())

print_area(Circle(2))
print_area(Square(3))
```

## 6. Metody i atrybuty klasowe

- Metoda instancyjna: pierwszy parametr `self` (dziaÅ‚a na instancji).
- `@classmethod`: pierwszy parametr `cls`, dostÄ™p do klasy, czÄ™sto jako alternatywne konstruktory.
- `@staticmethod`: nie przyjmuje `self`/`cls`; narzÄ™dziowa funkcja powiÄ…zana z klasÄ….
- Atrybut klasowy: wspÃ³Å‚dzielony przez wszystkie instancje.

ðŸ“˜ PrzykÅ‚ad: `Employee` z atrybutem `company_name` i metodÄ… `from_string("Jan,4000")`.


```
class Employee:
    company_name = "ACME"

    def __init__(self, name: str, salary: float):
        self.name = name
        self.salary = float(salary)

    @classmethod
    def from_string(cls, text: str) -> "Employee":
        # Alternatywny konstruktor z prostego formatu "Name,Salary"
        name, salary = [x.strip() for x in text.split(",", 1)]
        return cls(name, float(salary))

    @staticmethod
    def validate_salary(value: float) -> bool:
        # Nie uÅ¼ywa self/cls â€” czysta walidacja
        return value >= 0

emp = Employee.from_string("Jan, 4000")
print(emp.name, emp.salary, emp.company_name)
print("Walidacja wynagrodzenia 3000:", Employee.validate_salary(3000))
```

## 7. SkÅ‚adowanie obiektÃ³w i kompozycja

- **Dziedziczenie**: relacja "jest-rodzajem" (isâ€‘a). Np. `Dog` jest `Animal`.
- **Kompozycja**: relacja "ma" (hasâ€‘a). Obiekt zawiera inne obiekty i deleguje im odpowiedzialnoÅ›ci.

ðŸ“˜ PrzykÅ‚ad: `Order` zawierajÄ…cy listÄ™ `Product` i metodÄ™ `total_price()`.

ðŸ§© Zadanie: UtwÃ³rz klasy `Library`, `Book`, `Author`. `Library` ma listÄ™ `Book` i potrafi je wypisaÄ‡.


```
class Product:
    def __init__(self, name: str, price: float):
        self.name = name
        self.price = float(price)

class Order:
    def __init__(self):
        self.items: list[Product] = []  # kompozycja: Order "ma" Products

    def add(self, product: Product) -> None:
        self.items.append(product)

    def total_price(self) -> float:
        return sum(p.price for p in self.items)

order = Order()
order.add(Product("KsiÄ…Å¼ka", 39.99))
order.add(Product("Notes", 9.50))
print("Suma:", order.total_price())
```

## 8. Klasy abstrakcyjne i interfejsy

- ModuÅ‚ `abc` pozwala definiowaÄ‡ abstrakcyjne klasy bazowe (ABC).
- `@abstractmethod` wymusza implementacjÄ™ metody w klasach pochodnych.
- UÅ¼yteczne, by zdefiniowaÄ‡ kontrakt (interfejs) dla grupy klas.

ðŸ“˜ PrzykÅ‚ad: abstrakcyjna `Shape` z `area()` wymaganÄ… w `Circle`, `Rectangle`.

ðŸ§© Zadanie: Dodaj abstrakcyjnÄ… metodÄ™ `perimeter()` i zaimplementuj jÄ… w klasach pochodnych.


```
from abc import ABC, abstractmethod

class AShape(ABC):
    @abstractmethod
    def area(self) -> float:
        pass

    @abstractmethod
    def perimeter(self) -> float:
        pass

class ACircle(AShape):
    def __init__(self, r: float):
        self.r = r
    def area(self) -> float:
        return math.pi * self.r ** 2
    def perimeter(self) -> float:
        return 2 * math.pi * self.r

class ARectangle(AShape):
    def __init__(self, a: float, b: float):
        self.a = a
        self.b = b
    def area(self) -> float:
        return self.a * self.b
    def perimeter(self) -> float:
        return 2 * (self.a + self.b)

print(ACircle(2).area(), ACircle(2).perimeter())
print(ARectangle(2, 3).area(), ARectangle(2, 3).perimeter())
```

## 9. PrzeciÄ…Å¼anie operatorÃ³w (metody specjalne)

- `__str__`, `__repr__` â€” prezentacja obiektu (dla uÅ¼ytkownika i debugowania).
- `__add__`, `__sub__`, `__mul__`, `__eq__`, `__lt__`, ... â€” przeciÄ…Å¼anie operatorÃ³w.
- Pozwala sprawiÄ‡, by obiekty zachowywaÅ‚y siÄ™ jak typy wbudowane w okreÅ›lonych kontekstach.

ðŸ“˜ PrzykÅ‚ad: `Vector` z `__add__` i `__repr__`.

ðŸ§© Zadanie: Dodaj obsÅ‚ugÄ™ `==` (`__eq__`) i `len()` (`__len__`) dla `Vector`.


```
class Vector:
    def __init__(self, *coords: float):
        self.coords = tuple(coords)

    def __repr__(self) -> str:
        # Reprezentacja dla debugowania/REPL
        return f"Vector{self.coords!r}"

    def __add__(self, other: "Vector") -> "Vector":
        # Dodawanie wektorÃ³w o tej samej dÅ‚ugoÅ›ci
        if len(self.coords) != len(other.coords):
            raise ValueError("Wektory muszÄ… mieÄ‡ ten sam wymiar")
        return Vector(*[a + b for a, b in zip(self.coords, other.coords)])

    def __eq__(self, other: object) -> bool:
        if not isinstance(other, Vector):
            return NotImplemented
        return self.coords == other.coords

    def __len__(self) -> int:
        # dÅ‚ugoÅ›Ä‡ w sensie liczby skÅ‚adowych (nie norma)
        return len(self.coords)

v1 = Vector(1, 2, 3)
v2 = Vector(4, 5, 6)
print(v1 + v2)
print("RÃ³wne?", Vector(1, 2) == Vector(1, 2))
print("Wymiar v1:", len(v1))
```

## 10. Zaawansowane koncepcje OOP

- **Metaklasy**: klasy tworzÄ…ce klasy. DomyÅ›lna metaklasa to `type`. Rzadko potrzebne â€” krÃ³tka demonstracja.
- **MRO (Method Resolution Order)**: kolejnoÅ›Ä‡ wyszukiwania atrybutÃ³w/metod przy dziedziczeniu wielokrotnym (C3 linearization). SprawdÅº `Class.__mro__` lub `mro()`.
- **Mixin-y**: klasy dostarczajÄ…ce wycinek funkcjonalnoÅ›ci, uÅ¼ywane w wielokrotnym dziedziczeniu.
- **Dataclasses**: generujÄ… `__init__`, `__repr__`, `__eq__` dla â€žklas danychâ€.

ðŸ“˜ PrzykÅ‚ad: `@dataclass` z automatycznym `__init__` i `__repr__` oraz prosta metaklasa.

ðŸ§© Zadanie: UtwÃ³rz klasÄ™ `Point` jako `dataclass`, a nastÄ™pnie funkcjÄ™ liczÄ…cÄ… dystans Euklidesowy miÄ™dzy dwoma punktami.


```
# Metaklasa â€“ prosta demonstracja: logowanie tworzenia klas
class LoggingMeta(type):
    def __new__(mcls, name, bases, namespace):
        print(f"[LoggingMeta] Tworzenie klasy: {name}")
        return super().__new__(mcls, name, bases, namespace)

class LoggedExample(metaclass=LoggingMeta):
    pass

# Dataclass
from dataclasses import dataclass

@dataclass
class Point:
    x: float
    y: float

def distance(a: Point, b: Point) -> float:
    # Dystans Euklidesowy
    dx = a.x - b.x
    dy = a.y - b.y
    return (dx * dx + dy * dy) ** 0.5

p1 = Point(0, 0)
p2 = Point(3, 4)
print("Punkty:", p1, p2)
print("Dystans:", distance(p1, p2))
```
