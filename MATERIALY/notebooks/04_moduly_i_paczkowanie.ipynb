## 1. Wprowadzenie do moduÅ‚Ã³w

- ModuÅ‚ = plik `.py` zawierajÄ…cy kod Pythona (funkcje, klasy, staÅ‚e). MoÅ¼esz go importowaÄ‡ do innego pliku.
- Import: `import`, `from ... import ...`, aliasy `as`.
- RÃ³Å¼nica: import caÅ‚ego moduÅ‚u vs import wybranych nazw.
- Gdzie Python szuka moduÅ‚Ã³w? Lista katalogÃ³w w `sys.path` (bieÅ¼Ä…cy katalog, instalacje, site-packages itd.).


````
# PrzykÅ‚ady importu â€” korzystamy z wbudowanych moduÅ‚Ã³w dla demonstracji
import math  # import caÅ‚ego moduÅ‚u
import sys   # import caÅ‚ego moduÅ‚u

from math import sqrt, pi  # import wybranych nazw
import math as m            # alias moduÅ‚u

print(math.cos(0))      # uÅ¼ycie peÅ‚nej nazwy
print(sqrt(9))          # bez prefiksu (uwaÅ¼aj na kolizje nazw)
print(m.tau, pi)

# Gdzie Python szuka moduÅ‚Ã³w?
for p in sys.path[:5]:  # pokaÅ¼ pierwsze kilka wpisÃ³w
    print("sys.path ->", p)
````

### Uwaga o stylach importu

- `import math` â€” czytelne i bezpieczne; odwoÅ‚ujesz siÄ™ jako `math.cos`.
- `from math import sqrt` â€” wygodne, ale uwaÅ¼aj na kolizje nazw.
- `import x as y` â€” uÅ¼ywaj oszczÄ™dnie, tylko dla powszechnych aliasÃ³w (`import numpy as np`).


## 2. Tworzenie wÅ‚asnego moduÅ‚u

- WÅ‚asny moduÅ‚ to zwykÅ‚y plik `.py`.
- Przyklad: `math_utils.py` z funkcjami matematycznymi.
- UÅ¼ycie w innym pliku: `import math_utils` albo `from math_utils import area_circle`.
- Blok uruchomieniowy: `if __name__ == "__main__":` â€” kod wykonywany tylko przy uruchamianiu pliku jako programu.


````
# ZawartoÅ›Ä‡ przykÅ‚adowego pliku: math_utils.py (pokazujemy jako tekst)
math_utils_py = '''
# math_utils.py
import math

PI = math.pi  # staÅ‚a eksportowana

def area_circle(r: float) -> float:
    """Pole koÅ‚a o promieniu r."""
    return math.pi * r * r

def hypotenuse(a: float, b: float) -> float:
    """Twierdzenie Pitagorasa."""
    return math.hypot(a, b)

if __name__ == "__main__":
    # Prosty test moduÅ‚u uruchomionego bezpoÅ›rednio
    print("PI:", PI)
    print("area_circle(2):", area_circle(2))
    print("hypotenuse(3,4):", hypotenuse(3,4))
'''
print(math_utils_py)
````

### Jak uÅ¼yÄ‡ w innym pliku?

ZaÅ‚Ã³Å¼my, Å¼e masz w tym samym katalogu dwa pliki:

- `math_utils.py` (jak wyÅ¼ej)
- `main.py`

ZawartoÅ›Ä‡ `main.py` moÅ¼e wyglÄ…daÄ‡ tak:

```python

print(math_utils.PI)
print(area_circle(3))
```

UruchamiajÄ…c `python main.py`, zobaczysz wynik uÅ¼ycia funkcji z moduÅ‚u.


## 3. Paczki (packages)

- Paczka to katalog z plikami `.py`, ktÃ³ry zawiera plik `__init__.py` (choÄ‡ od Pythona 3.3 tzw. implicit namespace packages mogÄ… dziaÅ‚aÄ‡ bez niego, klasyczna paczka zwykle go ma).
- RÃ³Å¼nica: moduÅ‚ = pojedynczy plik `.py`; paczka = katalog moduÅ‚Ã³w.
- Importy wewnÄ…trz paczki: absolutne i wzglÄ™dne.
- PrzykÅ‚ad paczki `shapes/` z moduÅ‚ami `circle.py`, `square.py`.


````
# PrzykÅ‚adowa struktura paczki (jako tekst)
shapes_tree = '''
shapes/                # paczka
â”œâ”€â”€ __init__.py        # inicjalizacja paczki
â”œâ”€â”€ circle.py          # moduÅ‚
â””â”€â”€ square.py          # moduÅ‚
'''
print(shapes_tree)

# ZawartoÅ›Ä‡ plikÃ³w:
circle_py = '''
# shapes/circle.py
from math import pi

def area(r: float) -> float:
    return pi * r * r

def perimeter(r: float) -> float:
    return 2 * pi * r
'''

square_py = '''
# shapes/square.py

def area(a: float) -> float:
    return a * a

def perimeter(a: float) -> float:
    return 4 * a
'''

init_py = '''
# shapes/__init__.py
# Eksportujemy najwaÅ¼niejsze elementy na poziom paczki
from .circle import area as circle_area, perimeter as circle_perimeter
from .square import area as square_area, perimeter as square_perimeter

__all__ = [
    "circle_area", "circle_perimeter",
    "square_area", "square_perimeter",
]
'''

print(circle_py)
print(square_py)
print(init_py)
````

### Importy absolutne vs wzglÄ™dne

- Absolutne: `from shapes.circle import area` â€” zaczynajÄ… od korzenia pakietu.
- WzglÄ™dne (wewnÄ…trz paczki): `from .circle import area` lub `from ..utils import helper`.
- Zalecenie: preferuj absolutne importy dla czytelnoÅ›ci, wzglÄ™dne uÅ¼ywaj oszczÄ™dnie do importÃ³w wewnÄ…trz paczki.


## 4. `__init__.py` i eksport symboli

- `__init__.py` wykonywany przy imporcie paczki; moÅ¼na w nim inicjalizowaÄ‡ stan, re-eksportowaÄ‡ symbole, definiowaÄ‡ `__all__`.
- `__all__` kontroluje, co zostanie zaimportowane przy `from shapes import *`.
- Uwaga: `import *` w kodzie produkcyjnym jest niezalecane; uÅ¼ywaj jawnych importÃ³w.


````
# Demonstrujemy import selektywny (na podstawie definicji wyÅ¼ej w tekÅ›cie)
# Tu nie tworzymy faktycznych plikÃ³w, ale pokazujemy idiomatyczny kod:
example_use = '''
# main.py
from shapes import circle_area, square_perimeter

print(circle_area(1.5))
print(square_perimeter(4))
'''
print(example_use)
````

## 6. Organizacja wiÄ™kszego projektu

- Dziel projekt na logiczne moduÅ‚y i paczki: jedno zagadnienie = jeden moduÅ‚/paczka.
- Dobre praktyki:
  - Czytelne nazwy, snake_case dla moduÅ‚Ã³w/paczek.
  - Jasne API: eksportuj tylko to, co potrzebne (przez `__all__`, re-eksport w `__init__.py`).
  - Dokumentacja (docstringi), testy jednostkowe, formatowanie (Black), linting (Ruff/Flake8).

- `__main__.py`: pozwala uruchamiaÄ‡ paczkÄ™: `python -m mypkg`.


````
project_tree = '''
myproject/
â”œâ”€â”€ pyproject.toml
â”œâ”€â”€ src/
â”‚   â””â”€â”€ mypkg/
â”‚       â”œâ”€â”€ __init__.py
â”‚       â”œâ”€â”€ __main__.py      # punkt wejÅ›cia: python -m mypkg
â”‚       â”œâ”€â”€ core.py
â”‚       â””â”€â”€ utils.py
â””â”€â”€ tests/
    â””â”€â”€ test_core.py
'''
print(project_tree)

# PrzykÅ‚adowe __main__.py
main_py = '''
# src/mypkg/__main__.py
from .core import run

if __name__ == "__main__":
    run()
'''
print(main_py)
````

### ğŸ§© Zadanie 2: Paczka geometry
- StwÃ³rz paczkÄ™ `geometry` z moduÅ‚ami `circle.py` i `rectangle.py`.
- KaÅ¼dy moduÅ‚ ma funkcje `area(...)` i `perimeter(...)`.
- Dodaj `__init__.py`, reâ€‘eksportuj najwaÅ¼niejsze symbole.

Miejsce na Twoje rozwiÄ…zanie:


````
# ROZWIÄ„ZANIE â€” szkic
geometry_tree = '''
geometry/
â”œâ”€â”€ __init__.py
â”œâ”€â”€ circle.py
â””â”€â”€ rectangle.py
'''
geometry_init = '''
# geometry/__init__.py
from .circle import area as circle_area, perimeter as circle_perimeter
from .rectangle import area as rectangle_area, perimeter as rectangle_perimeter
__all__ = [
    "circle_area", "circle_perimeter",
    "rectangle_area", "rectangle_perimeter",
]
'''
geometry_circle = '''
# geometry/circle.py
from math import pi

def area(r: float) -> float: return pi * r * r

def perimeter(r: float) -> float: return 2 * pi * r
'''
geometry_rectangle = '''
# geometry/rectangle.py

def area(a: float, b: float) -> float: return a * b

def perimeter(a: float, b: float) -> float: return 2 * (a + b)
'''
demo_py = '''
# demo.py
from geometry import circle_area, rectangle_perimeter
print(circle_area(3))
print(rectangle_perimeter(2, 5))
'''
print(geometry_tree)
print(geometry_init)
print(geometry_circle)
print(geometry_rectangle)
print(demo_py)
````
